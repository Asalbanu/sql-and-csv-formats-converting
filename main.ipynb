```python
import pandas as pd
from sqlalchemy import create_engine
import sqlite3

df_csv = pd.read_csv('list_volcano.csv')

engine = create_engine('sqlite://',echo = False)
df_csv.to_sql('whole', con = engine, if_exists = 'append')

conn = sqlite3.connect("whole.db")
cursor = conn.cursor()


cursor = engine.execute("SELECT * FROM whole").fetchall()

conn.close()
cursor
```
